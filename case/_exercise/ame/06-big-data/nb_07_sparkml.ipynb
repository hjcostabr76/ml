```
import pyspark.pandas as ps
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.sql.functions import regexp_replace
```

```


df = ps.read_csv('/FileStore/tables/used_cars.csv')
df = df.to_spark()
display(df)
```

```
def enconding_categories(df, column_name):
    idx = StringIndexer(inputCol=column_name, outputCol=column_name+"Num")
    df_idx = idx.fit(df).transform(df)
    encoder = OneHotEncoder(inputCol=column_name+"Num", outputCol=column_name+"Vec")
    df_enc = encoder.fit(df_idx).transform(df_idx)
    return df_enc

df_enc = enconding_categories(df, 'fuel')
df_enc = enconding_categories(df_enc, 'seller_type')
df_enc = enconding_categories(df_enc, 'transmission')
df_enc = enconding_categories(df_enc, 'owner')
```

```
df_enc.show()
```

```
df_enc.count()
```

```
df_enc = df_enc.na.drop()
df_enc.count()
```

```
def clean_column (df, column_name, trash_string, number_type):
    return df_enc.withColumn(column_name+'_cln', regexp_replace(column_name, trash_string, "").cast(number_type))
```

```
df_enc = clean_column(df_enc, 'mileage', " kmpl", "double")
df_enc = clean_column(df_enc, 'engine', " CC", "int")
df_enc = clean_column(df_enc, 'max_power', " bhp", "int")
display(df_enc)
```

```
df_enc = df_enc.na.drop()
df_enc.count()
```

```
#coleque no inputCols apenas as variaveis que vc deseja usar na sua regressao
assembler = VectorAssembler(
        inputCols=['year', 'km_driven', 'fuelVec', 'seller_typeVec', 'transmissionVec', 
                  'ownerVec', 'mileage_cln', 'engine_cln', 'max_power_cln'],
        outputCol="features"
)

df_final = assembler.transform(df_enc)

display(df_final)
```

```
from pyspark.ml.feature import Normalizer

normalizer = Normalizer(inputCol="features", outputCol="normFeatures", p=1.0)
df_norm = normalizer.transform(df_final)
```

```
treino, teste = df_final.randomSplit([.8, .2], seed=42)

print(f'O Conjunto de treino tem: {treino.cache().count()} itens')
print(f'O Conjunto de teste tem: {teste.cache().count()} itens')

```

```
from pyspark.ml.regression import LinearRegression

reg = LinearRegression(featuresCol="features", labelCol="selling_price")

model = reg.fit(treino)

print(f'Ele intercepta em :{model.intercept:.2f}')
print(f'Ele tem os Coefs em :{model.coefficients[0]:.2f} e {model.coefficients[1]:.2f} e o coef 7 é {model.coefficients[6]:.2f}')

```

```
df_prev = model.transform(teste)
display(df_prev.select("features", 'selling_price', 'prediction'))
```

```
from pyspark.ml.evaluation import RegressionEvaluator

regEv = RegressionEvaluator(predictionCol="prediction", labelCol='selling_price', metricName="rmse")
rmse = regEv.evaluate(df_prev)

mse = regEv.setMetricName("mse").evaluate(df_prev)
r2 = regEv.setMetricName("r2").evaluate(df_prev)
mae = regEv.setMetricName("mae").evaluate(df_prev)

print(f"rmse: {rmse:0.2f}")
print(f"mse: {mse:0.2f}")
print(f"r2: {r2:0.2f}")
print(f"mae: {mae:0.2f}")
```
