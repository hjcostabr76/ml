### Proccess Data

```python
MAX_CATEG_VALUES = 4

cols_dual = []
cols_categorical = []
cols_continuous = []
```

```python
df_finance['Savings_objectives'] = df_finance['What are your savings objectives?']
df_finance = df_finance.drop(['What are your savings objectives?'], axis=1)
```

#### Categorize data

```python

for col in df_finance:
    n_unique = df_finance[col].nunique()
    print(f'"{col}": {df_finance[col].nunique()} unique values')
    if n_unique == 2:
        cols_dual.append(col)
    elif n_unique <= MAX_CATEG_VALUES:
        cols_categorical.append(col)

print('- ' * 15)
print('cols_dual: ', cols_dual)

print('- ' * 15)
print('cols_categorical: ', cols_categorical)
```

```python
for col in cols_dual:
    print(f'Unique values for "{col}":')
    print(df_finance[col].unique())
```

```python
df_finance['gender'] = np.where(df_finance['gender'] == 'Male', 1, 0).astype('uint8')
df_finance['Investment_Avenues'] = np.where(df_finance['Investment_Avenues'] == 'Yes', 1, 0).astype('uint8')
df_finance['Stock_Marktet'] = np.where(df_finance['Stock_Marktet'] == 'Yes', 1, 0).astype('uint8')


for col in cols_dual:
    print(f'Unique values for "{col}":')
    print(df_finance[col].unique())
```

```python
foo = pd.get_dummies(df_finance, columns=cols_categorical, drop_first=True)
# df_finance = pd.get_dummies(df_finance, columns=cols_categorical, drop_first=False)
# foo.columns
# foo.head()
# foo.info()
df_finance = foo
```

```python
df_finance.dtypes.unique()
# df_finance.select_dtypes(['0']).columns.tolist()
```

```python
cols_continuous = df_finance.select_dtypes(['int64']).columns.tolist()
cols_categorical = df_finance.select_dtypes(['uint8']).columns.tolist()
cols_categorical = list(filter(lambda x: x not in cols_dual, cols_categorical))

print('cols_continuous:')
print(cols_continuous)

print('- ' * 15)
print('cols_dual:')
print(cols_dual)

print('- ' * 15)
print('cols_categorical:')
print(cols_categorical)
```
