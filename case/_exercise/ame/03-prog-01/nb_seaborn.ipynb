## Seaborn x Matplotlib

Primeiramente, vamos recordar como fazer um gráfico de dispersão usando o __Matplotlib__:

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
plt.scatter('sepal_length', 'sepal_width', data = iris)

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Mostra o nosso Gráfico
plt.show()
```

No __Seaborn__, podemos fazer o mesmo gráfico utilizando a função __scatterplot__: 

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
sns.scatterplot(data = iris, x ='sepal_length', y= 'sepal_width')

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Mostra o nosso Gráfico
plt.show()
```

Mas então, a primeira pergunta é: __quais são as vantagens de usar o Seaborn ao invés do Matplotlib?__<br><br>
Vamos imaginar que queremos fazer o mesmo gráfico mas separando as pelos tipos de _Iris_ (ou seja, separar pelas classes):

```python
# Separe o DataFrame entre as 3 classes
df_ver = iris[iris['species'] == 'versicolor']
df_set = iris[iris['species'] == 'setosa']
df_vir = iris[iris['species'] == 'virginica']

# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
plt.scatter('sepal_length', 'sepal_width', data = df_ver)
plt.scatter('sepal_length', 'sepal_width', data = df_set)
plt.scatter('sepal_length', 'sepal_width', data = df_vir)

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Colocando a Legenda do Gráfico
plt.legend(['versicolor', 'setosa', 'virginica'])

# Mostra o nosso Gráfico
plt.show()
```

No __Seaborn__, sempre que quisermos quebrar um gráfico nas classes basta passar um parâmetro chamado __hue__:

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
sns.scatterplot(data = iris, x ='sepal_length', y= 'sepal_width', hue = 'species')

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Mostra o nosso Gráfico
plt.show()
```

E o __Seaborn__ nos permite customizar diversas detalhes do nosso gráfico:

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
sns.scatterplot(data = iris, 
                x ='sepal_length',
                y= 'sepal_width',
                hue = 'species',
                hue_order = ['versicolor', 'setosa', 'virginica'])

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Mostra o nosso Gráfico
plt.show()
```

```python
# Define a semente randômica
np.random.seed(42)

# Cria um valor de size no DataFrame
iris['size'] = np.random.randint(100, 800, iris.shape[0])

# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Escolhendo um Pallete de Cores
palette = ['#FF6978', '#B1EDE8', '#6D435A']

# Passando um parâmetro de Size para o gráfico
size = [500, 150, 300]

# Gráfico de Linha
sns.scatterplot(data = iris
                ,x ='sepal_length'
                ,y= 'sepal_width'
                ,hue = 'species'
                ,palette = palette
                ,hue_order = ['versicolor', 'setosa', 'virginica']
                ,s = iris['size']
               )

# Coloca o titulo
plt.title('Comprimento pela Largura da Sépala', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Comprimento', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Largura', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando Grid no Gráfico
plt.grid()

# Mostra o nosso Gráfico
plt.show()
```

### HeatMap

O __HeatMap__ (Mapa de Calor) é um gráfico bem interessante quando queremos criar um _Grid_ de variações.<br><br>
Para usar o HeatMap, vamos definir primeiramente as correlações do DataFrame _Iris_:

```python
corr = iris.corr()
corr
```

A função __corr__ mostra como as variáveis do DataFrame estão relacionadas, ou seja o quanto cada variável influência na variação das outras:

```python
# Cria o HeatMap
sns.heatmap(corr)

# Mostra o Gráfico
plt.show()
```

## Configurações com o Seaborn

Vamos aprender algumas outra modificações possíveis para fazermos com os gráficos:

```python
# Criando um DataFrame com dados simples
df = pd.DataFrame(data = [['caso 1',40, '40%'], ['caso 2', 30, '30%'], ['caso 3', 80, '80%']])

# Nomeando as colunas
df.columns = ['caso', 'valor', 'porcentagem']

# Visualiza o DataFrame
df
```

```python
# Cria o subplot
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(12,6))

# Cria o gráfico sem modificação
sns.barplot(x = df['caso'].values, 
            y = df['valor'].values, 
            ax = ax[0], 
            color = '#C3514E')

# Cria o gráfico que vamos modificar
sns.barplot(x = df['caso'].values, 
            y = df['valor'].values, 
            ax = ax[1], 
            color = '#0C8040')

# Removendo as linhas de Borda do Barplot
ax[1].spines['right'].set_visible(False)
ax[1].spines['bottom'].set_visible(False)
ax[1].spines['left'].set_visible(False)
ax[1].spines['top'].set_visible(False)

# Define um Titulo para o Gráfico do Seaborn
ax[1].set_title('Grafico Customizado\n',  weight ='bold')

# Colocando o Texto do Valor do Gráfico
ax[1].text(0, 42, '40%' , fontsize=11, color='black',verticalalignment='center',horizontalalignment='center')
ax[1].text(1, 32, '30%' , fontsize=11, color='black',verticalalignment='center',horizontalalignment='center')  
ax[1].text(2, 82, '80%' , fontsize=11, color='black',verticalalignment='center',horizontalalignment='center') 

# Remove a Grade da Esquerda
ax[1].tick_params(axis= 'both', bottom=True, labelbottom=True, left = False, labelleft= False)

# Mostra o Gráfico
plt.show()
```
