## Matplotlib

O **Matplotlib** é uma das principais bilbliotecas para o Python no quesito de visualização de Dados!<br><br>
Com esta biblioteca e com o auxílio de dados, conseguimos criar visualizações gráficas. Vamos começar por carregar a nossa biblioteca no notebook:

```python
# Carregando a biblioteca
import matplotlib.pyplot as plt
```

Agora vamos passar por alguns tipos de gráficos mais utilizados do Matplotlib:

### Gráfico de Barras

```python
# Vamos criar as nossas categorias
alunos = ['Maria Eduarda', 'José Eduardo', 'Ana Paula', 'Rafael', 'Pedro']

# Agora atribuir um valor de nota para cada um deles
notas = [9.5, 8, 7.5, 10, 6]
```

```python
# Criando um gráfico de Barras
plt.bar(alunos, notas)
plt.show()
```

Criamos o nosso primeiro gráfico de barras, mas faltam algumas informações como título do gráfico e eixos, além disso está acontecendo um problema nos nomes dos alunos. Vamos ver como ajustar esses pontos:

```python
# cria o nosso gráfico de barras
plt.bar(alunos, notas)

# Colocando um titulo no gráfico
plt.title('Notas dos Alunos da Lets Code')

# Colocando nome nos eixos agora
plt.xlabel('Nome dos Alunos')
plt.ylabel('Notas')

# Ajustando os nomes de acordo com um certo ângulo
plt.xticks(rotation=45)
plt.show()
```

Vamos colocar mais laguns toques estéticos no nosso gráfico:<br><br>

- Aumentar a fonte do título;
- Aumentar a fonte do título dos eixos;
- Aumentar a fonte do nomes dos alunos;
- Mudar a cor do gráfico, para verde por exemplo;
- Aumentar o tamanho desse gráfico!<br><br>

Vamos ver como fazer isso então:

```python
# vamos definir um tamanho para a nossa figura
plt.figure(figsize = (12,8))

# cria o nosso gráfico de barras, passando um valor para color
plt.bar(alunos, notas, color = 'green')

# Colocando um titulo no gráfico
plt.title('Notas dos Alunos da Lets Code', fontsize = 15)

# Colocando nome nos eixos agora
plt.xlabel('Nome dos Alunos', fontsize = 15)
plt.ylabel('Notas', fontsize = 15)

# Ajustando os nomes de acordo com um certo ângulo
plt.xticks(rotation=45, fontsize = 12)
plt.yticks(fontsize = 12)

plt.show()
```

Vamos agora mudar a cor novamente, dessa vez para azul e vamos aproveitar melhor o tamanho do gráfico no nosso notebook. Por fim vamos salvar essa figura em um arquivo png externo:

```python
# vamos definir um tamanho para a nossa figura
plt.figure(figsize = (12,8))

# cria o nosso gráfico de barras, passando um valor para color
plt.bar(alunos, notas, color = 'blue')

# Colocando um titulo no gráfico
plt.title('Notas dos Alunos da Lets Code', fontsize = 15)

# Colocando nome nos eixos agora
plt.xlabel('Nome dos Alunos', fontsize = 15)
plt.ylabel('Notas', fontsize = 15)

# Ajustando os nomes de acordo com um certo ângulo
plt.xticks(rotation=45, fontsize = 12)
plt.yticks(fontsize = 12)

#Ajustando o Plot a tela
plt.tight_layout()

#Mostra o nosso gráfico
plt.show()

#Salva a figura em um arquivo png
plt.savefig("barras_azul.png")
```

### Gráfico de setores (Famoso Gráfico de Pizza)

```python
cursos = ['Python', 'DS', 'DS & AI', 'Python for Finance', 'Banco de Dados']
matriculados = [308, 127, 200, 85, 50]
```

```python
plt.figure(figsize = (8,8))

# gráfico de setor, com as labels
plt.pie(matriculados, labels = cursos, autopct='%.2f%%')

plt.title('Matrículas em cursos de programação', fontsize = 12)

# deixando o gráfico redondinho
plt.axis('equal')

plt.tight_layout()
plt.show()
```

### Gráfico de Linhas

```python
dados_x = [0, 1, 2, 3, 4]
dados_y = [1, 3, 5, 7, 9]
```

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (10,8))

# Gráfico de Linha
plt.plot(dados_x, dados_y)

# Coloca o titulo
plt.title('Gráfico de Linha', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Eixo X', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Eixo Y', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Mostra o nosso Gráfico
plt.show()
```

```python
dados_x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
dados_y = [0, 1, 2, 4, 8, 16, 32, 64, 128]
```

Assim como no Gráfico de barras, podemos customizar marcadores, cor e tipos de linha para o gráfico:

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (10,8))

# Gráfico de Linha
plt.plot(dados_x, dados_y, marker='x', color='red', ls='--')

# Coloca o titulo
plt.title('Parábola', fontsize = 15)

# Titulo para eixo X
plt.xlabel('x', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('y = x²', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Mostra o nosso Gráfico
plt.show()
```

Conseguimos também colocar dois conjuntos de dados em um mesmo gráfico:

```python
# Primeiro conjunto de dados
dados_x1 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
dados_y1 = [5, 10, 15, 20, 25, 30, 35, 40, 45]

# Segundo conjunto de dados
dados_x2 = [0, 1, 2, 3, 4, 5, 6, 7, 8]
dados_y2 = [0, 1, 4, 9, 16, 25, 36, 49, 64]
```

```python
# Ajustando o tamanho da Figura
plt.figure(figsize = (10,8))

# Gráfico do Primeiro conjunto
plt.plot(dados_x1, dados_y1, marker='o', color='blue', ls='-')

# gráfico do Segundo Conjunto
plt.plot(dados_x2, dados_y2, marker='x', color='green', ls='--')

# Coloca o titulo
plt.title('Várias Curvas', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Eixo X', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Eixo Y', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Colocando uma legenda no nosso gráfico
plt.legend(['y = 5x+5', 'y = x²'])

# Mostra o nosso Gráfico
plt.show()
```

### Subplots

```python
dados_x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
# e 4 conjuntos diferentes para a coordenada Y
dados_y1 = [5, 10, 15, 20, 25, 30, 35, 40, 45]
dados_y2 = [0, 1, 2, 4, 8, 16, 32, 64, 128]
dados_y3 = [0.0, 0.3131151331286599, 0.5947403868365128, 0.8165528078619212, 0.9562447986054508, 0.9997675844179418, 0.9427440939531498, 0.7909091597020744, 0.5595327678952049]
dados_y4 = [1.0, 0.9497151748844606, 0.8039178268116431, 0.5772707440818498, 0.29256774453111023, -0.02155869075601868, -0.333516976054375, -0.6119335757248155, -0.828808229719958]

plt.figure(figsize=(12,12))

# 2 linhas, 2 colunas, estamos no gráfico 1
plt.subplot(221)
plt.plot(dados_x, dados_y1, marker='o', color='blue', ls='-')
plt.title('Gráfico 1')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# 2 linhas, 2 colunas, estamos no gráfico 2
plt.subplot(222)
plt.plot(dados_x, dados_y2, marker='x', color='green', ls='--')
plt.title('Gráfico 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# 2 linhas, 2 colunas, estamos no gráfico 3
plt.subplot(223)
plt.plot(dados_x, dados_y3, marker='+', color='red', ls=':')
plt.title('Gráfico 3')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# 2 linhas, 2 colunas, estamos no gráfico 4
plt.subplot(224)
plt.plot(dados_x, dados_y4, marker='s', color='purple', ls='-.')
plt.title('Gráfico 4')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```

##   

E também podemos juntar alguns gráficos nos subplots:

```python
dados_x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
# 4 conjuntos de coordenadas Y
dados_y1 = [5, 10, 15, 20, 25, 30, 35, 40, 45]
dados_y2 = [0, 1, 2, 4, 8, 16, 32, 64, 128]
dados_y3 = [0.0, 0.3131151331286599, 0.5947403868365128, 0.8165528078619212, 0.9562447986054508, 0.9997675844179418, 0.9427440939531498, 0.7909091597020744, 0.5595327678952049]
dados_y4 = [1.0, 0.9497151748844606, 0.8039178268116431, 0.5772707440818498, 0.29256774453111023, -0.02155869075601868, -0.333516976054375, -0.6119335757248155, -0.828808229719958]


# 2 linhas, 1 coluna, estamos no gráfico 3
# note que falamos para o Python que estamos no gráfico 2
# isso ocorre porque estamos afirmando que teremos 2 linhas de 1 coluna cada
# o que será o nosso gráfico 3, por hora, é visto pelo Python como gráfico 2

plt.figure(figsize=(12,12))
plt.subplot(212)
plt.plot(dados_x, dados_y3, marker='+', color='red', ls=':')
plt.plot(dados_x, dados_y4, marker='s', color='purple', ls='-.')
plt.title('Gráfico 3')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(221)
plt.plot(dados_x, dados_y1, marker='o', color='blue', ls='-')
plt.title('Gráfico 1')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# 2 linhas, 2 colunas, estamos no gráfico 2
# ué, mas já não havíamos usado o gráfico 2 antes?
# agora que redefinimos o número de colunas para 2, o Python considera a posição 2
# como sendo a segunda coluna da primeira linha...
# nosso antigo gráfico 2 agora é visto pelo Python como gráfico 3
plt.subplot(222)
plt.plot(dados_x, dados_y2, marker='x', color='orange', ls='--')
plt.title('Gráfico 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

Vamos agora criar um gráfico de dispersão a partir de um DataFrame:

```python
import pandas as pd

df = pd.read_csv('./datasets/IBOV.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.fillna(method='ffill')
```

```python
df_1trim = df[(df['Date'] >= '2020-01-01') & (df['Date'] < '2020-04-01')]
df_1trim
```

```python
# Ajustando o tamanho da figura
plt.figure(figsize = (15, 10))

# Criando um gráfico de dispersão para as máximas
plt.scatter('Date', 'High', data = df_1trim)

# Criando um gráfico de dispersão para as minimas
plt.scatter('Date', 'Low', data = df_1trim)

# Criano um gráfico de linha para o valor de fechamento
plt.plot('Date', 'Close', data = df_1trim)

# Coloca o titulo
plt.title('Bovespa', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Data', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Fechamento', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(rotation = 90, fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Legenda do Gráfico
plt.legend(["Fechamento", "Max.", "Min"]) # adicionamos uma legenda para facilitar para nosso leitor

# Mostra o gráfico
plt.show()
```
