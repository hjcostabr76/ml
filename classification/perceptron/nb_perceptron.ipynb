### Imports

```python
import numpy as np
import matplotlib.pyplot as plt

import math_utils
from Perceptron import Perceptron
```

## Generate Sample

```python
inputsCount = 4 # ?
N = 50
randomAvgs = np.ceil(np.random.uniform(0, 50, inputsCount))
randomStdDev = np.ceil(np.random.uniform(1, 3, inputsCount))

c1x1 = np.random.normal(randomAvgs[0], randomStdDev[0], N)
c1x1 = np.transpose([c1x1])
c1x2 = np.random.normal(randomAvgs[0], randomStdDev[1], N)
c1x2 = np.transpose([c1x2])

X1 = np.hstack((c1x1, c1x2, np.ones((N, 1))))

c2x1 = np.random.normal(randomAvgs[1], randomStdDev[0], N)
c2x1 = np.transpose([c2x1])
c2x2 = np.random.normal(randomAvgs[1], randomStdDev[1], N)
c2x2 = np.transpose([c2x2])

X2 = np.hstack((c2x1, c2x2, np.zeros((N, 1))))
X = np.vstack((X1, X2))

# X


# separationLineYs = MathUtils.getFOfX(rangeTestPoints, coeficients)


plt.plot(X1[:, 0], X1[:, 1], '.r')
plt.plot(X2[:, 0], X2[:, 1], '.g')
# # plt.plot(x3, x4, '.b')
# # plt.plot(rangeTestPoints, separationLineYs, '-k')
# # plt.plot(rangeTestPoints, rangeTestPoints, '-k')

# # Config

# plt.rcParams["figure.figsize"] = [12, 6]
# plt.title('Samples: Classes 01 & 02')
# plt.legend(['Class 01', 'Class 02'])
# # plt.xlabel('x1')
# # plt.ylabel('x2')

# # plt.savefig('graph-01-separation-line.png')
# plt.show()
```

# Train

```python
perceptron = Perceptron(X[:, :-1], X[:, -1], tolerance=0.0001, step=.5, maxIterations=1000, actvFunc='step')
w = perceptron.train()
```

# Test

```python


# Montar linha de separacao dos grupos
minX1 = np.min(X[:, 0])
maxX1 = np.max(X[:, 0])
padding = 10
T = np.linspace(minX1 - padding, maxX1 + padding, 100)
yApprox = list(map(lambda t: math_utils.linFunc(t, w), T))

# Montar grafico de posicao dos pontos

T2 = np.linspace(0, X.shape[0] - 1, X.shape[0])
bar = list(map(lambda x: x[-1], X))
baz = list(map(lambda x: perceptron.getPrediction(x[:-1]), X))

# Exibir
plt.rcParams["figure.figsize"] = [30, 15]

plt.plot(X1[:, 0], X1[:, 1], '.r')
plt.plot(X2[:, 0], X2[:, 1], '.g')
plt.plot(T, yApprox, '-w')
plt.show()

plt.plot(T2, bar, 'xr')
plt.plot(T2, baz, '.w')
```
