```python
def pca(chico: np.array) -> None:

    foo = chico[:, 0:1]
    y = chico[:, -1]

    mean = np.mean(foo, 0)
    bar = foo - mean

    cov = np.cov(bar.T)
    eigValue, eigVector = LA.eig(cov)

    print(eigValue)
    print(eigVector)

    c1 = lambda x: eigVector[0, 0]*x + eigVector[1, 0]
    c2 = lambda x: eigVector[0, 1]*x + eigVector[1, 1]

    baz = np.linspace(-3, 3, 50)

    plt.figure(figsize=(12, 10))
    plt.grid(alpha=.3)
    
    plt.plot(baz, c1(baz), 'r:')
    plt.plot(baz, c2(baz), 'g:')
    plt.scatter(bar[:, 0], bar[:, 1])
```
